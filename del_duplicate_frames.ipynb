```python
import os
from PIL import Image
import imagehash

# Function to calculate perceptual hash for an image
def calculate_hash(image_path):
    img = Image.open(image_path)
    return imagehash.average_hash(img)  # You can also try dhash, phash, whash

# Function to identify and remove similar images
def remove_similar_images(directory, hash_size=8, threshold=5):
    images = os.listdir(directory)
    unique_images = {}
    duplicates = []

    for img_name in images:
        img_path = os.path.join(directory, img_name)
        if os.path.isfile(img_path):
            img_hash = calculate_hash(img_path)
            
            # Compare with existing images in the unique set
            is_duplicate = False
            for existing_hash in unique_images.values():
                if img_hash - existing_hash < threshold:  # Similar if hash difference is less than threshold
                    is_duplicate = True
                    duplicates.append(img_name)
                    break
            
            if not is_duplicate:
                unique_images[img_name] = img_hash

    # Removing duplicates
    for duplicate in duplicates:
        os.remove(os.path.join(directory, duplicate))
        print(f"Removed duplicate: {duplicate}")

    print(f"Remaining unique images: {len(unique_images)}")

# Directory containing images
image_directory = 'File 3/frames_workouts3'

# Call the function
remove_similar_images(image_directory)
```
